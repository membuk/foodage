```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
path = "/content/Veriler.csv"
df = pd.read_csv(path,sep=";")
data = df.iloc[:-1,1:].values
data = np.transpose(data)
total = df.iloc[-1][1:].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, total, test_size = 1/5, random_state = 123, shuffle=1)
print(X_train.shape)
model = LinearRegression()
model.fit(X_train, y_train)
predict = model.predict(X_test)
print("Pre \t","Real \t","gap")
for i in range(len(X_test)):
  print(int(predict[i]),"\t",y_test[i],"\t",abs(int(predict[i])-y_test[i]))
```

```python
coefs = model.coef_
intercept = model.intercept_
print(coefs,intercept)
print(model.get_params())
```

```python
model.score(X_test,y_test)
```
